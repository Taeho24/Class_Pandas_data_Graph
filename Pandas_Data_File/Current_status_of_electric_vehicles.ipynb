```python
# pandas
import pandas as pd
import matplotlib.pyplot as plt


df4 = pd.read_excel('./한국전력공사_지역별 전기차 현황정보_20230331.xlsx',  header=0)
df4=df4.transpose()
df4.head(20)
```

```python
df = df4.loc['합계']

plt.style.use('ggplot')

plt.plot(df.index, df.values, marker='o', markersize=5)

plt.title(' Current status of electric vehicles')
plt.xlabel('time')
plt.ylabel('among')

plt.show()
```
