```python
# pandas
import pandas as pd
import matplotlib.pyplot as plt

df2 = pd.read_excel('./한국전력공사_지역별 전기차 충전기 현황정보_20221231.xlsx')
df3 = pd.read_excel('./한국전력공사_지역별 전기차 충전소 현황정보_20230718.xlsx')
#지역 이름이 두번 중복으로 나오는 부분 수정 필요
df= df2+df3

print(df.head())

plt.style.use('ggplot')
df.plot(kind='bar', figsize=(10,20), width=0.7)

plt.title('Number of charging stations')
plt.xlabel('Local')
plt.ylabel('Amount')

plt.show()
```
